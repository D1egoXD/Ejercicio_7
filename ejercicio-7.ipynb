#### b). Construye un diagrama de barras y uno circular.

## Diagrama de Barras

```python
plt.figure(figsize = (12, 6)) # (ancho, largo) de la grafica

marcas_clase = [1, 2, 3, 4, 5, 6] # datos del eje x
frecuencia = [8, 8, 5, 4, 3, 2] # datos del eje y
marcas_texto = ["Pumas", "América", "Chivas", "Toluca", "Santos", "Necaxa"]

# Diagramas de Barras con:
# - barras ajestadas al 80% (height = 0.8)
# - contorno negro (edgecolor = k)
plt.barh(marcas_clase, frecuencia, 
        height = 0.8, edgecolor = "k", 
        color = ["#ECA520", "#FCF906", "#E74C3C", "#F90F0B", "#2ABF10", "#ED4C29"]) 


plt.yticks(marcas_clase, marcas_texto, fontsize = 15, rotation = 20)
plt.xlabel("Frecuancia Absoluta", fontsize = 25) # etiqueta del eje x
plt.ylabel("Marcas de Clase", fontsize = 25) # etiqueta del eje y
plt.title("Diagramas de Barras", fontsize = 40) # titulo de la grafica
plt.grid() # se activa la cuadricula
plt.show() # se muestra el grafico en pantalla
```

## Diagrama de Pastel

```python
datos = [0.2667, 0.2667, 0.1587, 0.1333, 0.1000, 0.0667]
separaciones = [0, 0, 0, 0, 0, 0]

plt.pie(datos,
       explode = separaciones,
       labels = ["Pumas","America", "Chivas", "Toluca", "Santos", "Necaxa"],
       colors = ["#ECA520", "#FCF906", "#E74C3C", "#F90F0B", "#2ABF10", "#ED4C29"],
       autopct = "%0.01f%%",
       pctdistance = 0.8,
       counterclock = False,
       startangle = 89)
plt.show()
```
